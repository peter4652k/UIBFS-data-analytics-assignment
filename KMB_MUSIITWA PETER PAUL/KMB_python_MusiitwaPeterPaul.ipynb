```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.style.use('seaborn-v0_8-whitegrid')

#global chart parameters for consistency
plt.rcParams.update({
    'axes.titlesize': 14,
    'axes.titleweight': 'bold',
    'axes.labelweight': 'bold',
    'figure.figsize': (10, 6)
})

#  KMB Bank Brand Colors
KMB_NAVY = '#1F3864'
KMB_GOLD = '#C9A84C'

print("Libraries imported and style configured successfully.")
```

```python
# Load the dataset
loans = pd.read_excel('KMB_Loan_Portfolio_Dataset.xlsx')


# Verify the shape
print(f"Loans shape: {loans.shape}")
loans.head()
```

```python
# Load the specific sheet 'Loan_Applications' from the dataset
df = pd.read_excel('KMB_Loan_Portfolio_Dataset.xlsx', sheet_name='Loan_Applications')

```

```python
# 1. Display the first 10 rows
display(df.head(10))

```

```python
# Profiling commands
print("--- Data Info ---")
df.info()

print("\n--- Statistical Summary ---")
display(df.describe())

```

```python
# missing values per column
print("\n--- Missing Values ---")
missing_counts = df.isnull().sum()
print(missing_counts[missing_counts > 0])

```

```python
# Check for duplicate Loan_IDs
duplicates = df.duplicated(subset='Loan_ID').sum()
print(f"\nDuplicate Loan_IDs found: {duplicates}")

```

```python
# unique values in Loan_Status
unique_status = df['Loan_Status'].unique()
print(f"\nUnique Loan Status values: {unique_status}")
```

##### Task F2: Data Cleaning in Python 

```python
# Strip whitespace from all text (object) columns
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
#  Standardise Loan_Status to Title Case
df['Loan_Status'] = df['Loan_Status'].str.title()
```

```python
# Parse Disbursement_Date

df['Disbursement_Date'] = pd.to_datetime(df['Disbursement_Date'], dayfirst=True, errors='coerce')
failed_dates = df[df['Disbursement_Date'].isna()]
print(f"Rows with failed date parsing: {len(failed_dates)}")
```

```python
#  Convert Interest_Rate to Float

df['Interest_Rate'] = df['Interest_Rate'].str.replace('%', '', regex=False).astype(float)
```

```python
# Handle negative values in Amount_Paid
if 'Amount_Paid' in df.columns:
    df['Amount_Paid'] = df['Amount_Paid'].abs()
```

```python
# Fill specific missing values
fill_values = {
    'Loan_Officer': 'Unassigned',
    'Collateral_Type': 'None Provided',
    'Sector': 'Unclassified'
}
df.fillna(value=fill_values, inplace=True)
```

```python
# Verification
print("\n--- Verification: Missing Values After Cleaning ---")
print(df.isnull().sum())

print("\n--- Verification: Statistical Summary After Cleaning ---")
display(df.describe())
```

```python
import matplotlib.ticker as mtick

# Group and Sort the data
# sum the outstanding balance per sector and sort from highest to lowest
sector_balance = df.groupby('Sector')['Outstanding_Balance_UGX'].sum().sort_values(ascending=False).reset_index()

# Create the Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=sector_balance, 
    x='Sector', 
    y='Outstanding_Balance_UGX', 
    color=KMB_NAVY  # Use the KMB Navy color defined in F1
)

# Add Titles and Labels
plt.title('Outstanding Loan Balance by Sector', fontsize=16, pad=20)
plt.xlabel('Economic Sector', fontsize=12)
plt.ylabel('Total Outstanding Balance (UGX)', fontsize=12)
plt.xticks(rotation=45)  # Rotate labels to prevent overlap

# Professional UGX Formatting on Y-axis

fmt = '{x:,.0f} UGX'
tick = mtick.StrMethodFormatter(fmt)
plt.gca().yaxis.set_major_formatter(tick)

# Add a subtle grid and clean up layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
# Prepare the data
status_counts = df['Loan_Status'].value_counts()


colors = sns.color_palette('pastel')[0:len(status_counts)]

# 3. Create the Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(
    status_counts, 
    labels=status_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors,
    pctdistance=0.85, # Moves percentages slightly out for readability
    explode=[0.05 if i == 0 else 0 for i in range(len(status_counts))] 
)


centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# 5. Add Titles and Legend
plt.title('Distribution of Loans by Status', fontsize=16, pad=20)
plt.legend(status_counts.index, title="Loan Status", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.axis('equal')  
plt.tight_layout()
plt.show()
```

```python

# Calculate the mean for the vertical line
mean_principal = df['Principal_Amount_UGX'].mean()

# Create the Histogram
plt.figure(figsize=(12, 6))
sns.histplot(
    df['Principal_Amount_UGX'], 
    kde=True,          
    color=KMB_NAVY, 
    log_scale=True     
)

# Add the Vertical Mean Line
plt.axvline(
    mean_principal, 
    color=KMB_GOLD, 
    linestyle='--', 
    linewidth=2, 
    label=f'Mean: {mean_principal:,.0f} UGX'
)

# Add Titles and Labels
plt.title('Distribution of Principal Loan Amounts (Log Scale)', fontsize=16, pad=20)
plt.xlabel('Principal Amount (UGX) - Logarithmic Scale', fontsize=12)
plt.ylabel('Frequency (Number of Loans)', fontsize=12)
plt.legend()


from matplotlib.ticker import ScalarFormatter
plt.gca().xaxis.set_major_formatter(ScalarFormatter())
plt.ticklabel_format(style='plain', axis='x')

plt.tight_layout()
plt.show()
```

```python
# Create the Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df, 
    x='Loan_Status', 
    y='Interest_Rate', 
    palette='Set2'
)

#  Add Titles and Labels
plt.title('Interest Rate Distribution by Loan Status', fontsize=16, pad=20)
plt.xlabel('Loan Status', fontsize=12)
plt.ylabel('Interest Rate (%)', fontsize=12)

# Add a horizontal line for the overall median interest rate
overall_median = df['Interest_Rate'].median()
plt.axhline(overall_median, color='red', linestyle='--', alpha=0.5, label=f'Overall Median: {overall_median}%')

#  Clean up formatting
plt.legend()
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.tight_layout()

plt.show()
```

```python
# Create the Pivot Table using your exact column names

heatmap_data = df.pivot_table(
    index='Branch_Code', 
    columns='Sector', 
    values='Loan_ID', 
    aggfunc='count'
).fillna(0)

#  Create the Heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(
    heatmap_data, 
    annot=True,       
    fmt='.0f',       
    cmap='YlGnBu',    
    cbar_kws={'label': 'Number of Loans'}
)

# Add Titles and Labels
plt.title('Concentration Risk: Loan Count by Branch Code and Sector', fontsize=16, pad=20)
plt.xlabel('Economic Sector', fontsize=12)
plt.ylabel('Branch Code', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Prepare Time Series Data
# Ensure date is datetime and create a Year-Month period
df['Disbursement_Month'] = df['Disbursement_Date'].dt.to_period('M').astype(str)

# Group by month to get Count and Sum
monthly_data = df.groupby('Disbursement_Month').agg({
    'Loan_ID': 'count',
    'Principal_Amount_UGX': 'sum'
}).sort_index()

# Create the Dual Y-Axis Plot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Monthly Loan Counts 
color_count = KMB_GOLD
ax1.set_xlabel('Disbursement Month', fontsize=12)
ax1.set_ylabel('Number of Loans Issued', color=color_count, fontsize=12, fontweight='bold')
sns.barplot(x=monthly_data.index, y=monthly_data['Loan_ID'], ax=ax1, color=color_count, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color_count)

# Create a second y-axis for the Total Principal
ax2 = ax1.twinx()

# Total Principal Disbursed (Line Chart)
color_amt = KMB_NAVY
ax2.set_ylabel('Total Principal (UGX)', color=color_amt, fontsize=12, fontweight='bold')
sns.lineplot(x=range(len(monthly_data)), y=monthly_data['Principal_Amount_UGX'], 
             ax=ax2, color=color_amt, marker='o', linewidth=3)
ax2.tick_params(axis='y', labelcolor=color_amt)

# Professional UGX Formatting for the second axis
import matplotlib.ticker as mtick
fmt = '{x:,.0f} UGX'
tick = mtick.StrMethodFormatter(fmt)
ax2.yaxis.set_major_formatter(tick)


plt.title('Monthly Disbursement Trends: Loan Volume vs. Total Value', fontsize=16, pad=20)
ax1.set_xticklabels(monthly_data.index, rotation=45)
fig.tight_layout()

plt.show()
```
